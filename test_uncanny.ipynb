```python

x = np.arange(0., 7., 0.1)
y = np.arange(0., 7., 0.1)

X, Y = np.meshgrid(x, y)

def uncanny_function(t):
    return 1/(1 + np.exp(-(t-4)) ) - 1.2 * np.exp(-(t-6) ** 2)

def group_salience_function(t):
    return 1/(1 + np.exp(-(t-4)) ) - 1.5 * np.exp((t-8))

plt.figure()

Z = uncanny_function(X) + group_salience_function(Y)

ax = plt.axes(projection="3d")

ax.plot_wireframe(X, Y, Z)
ax.set_xlabel('Human Likeness')
ax.set_ylabel('Group Salience')
ax.set_zlabel('Attitude toward out-group')


ax.xaxis.set_tick_params(labelsize=0)
ax.zaxis.set_tick_params(labelsize=0)
ax.yaxis.set_tick_params(labelsize=0)


plt.show()
```

```python
x = np.arange(0., 7., 0.1)
y = np.arange(0., 7., 0.1)

X, Y = np.meshgrid(x, y)

def uncanny_function(t):
    return 1/(1 + np.exp(-(t-4)) ) - 1.2 * np.exp(-(t-6) ** 2)

def avatar_function(t):
    return t / 8.

plt.figure()

Z = uncanny_function(X) * avatar_function(Y)

ax = plt.axes(projection="3d")

ax.plot_wireframe(X, Y, Z)
ax.set_xlabel('Human Likeness')
ax.set_ylabel('Co-presence')
ax.set_zlabel('Attitude toward out-group member')


ax.xaxis.set_tick_params(labelsize=0)
ax.zaxis.set_tick_params(labelsize=0)
ax.yaxis.set_tick_params(labelsize=0)


plt.show()
```
